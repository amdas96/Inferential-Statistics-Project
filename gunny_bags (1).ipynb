## Business Context

The cement company utilizes gunny bags for packaging its products.

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter.

## Objective

The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain:

1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
#import the important packages
import pandas as pd #library used for data manipulation and analysis
import numpy as np # library used for working with arrays.
import matplotlib.pyplot as plt # library for plots and visualisations
import seaborn as sns # library for visualisations
%matplotlib inline 

import scipy.stats as stats# this library contains a large number of probability distributions as well as a growing library of statistical functions.
from scipy.stats import norm
```

##### Mean and Standard deviation:

```python
mean=5
std=1.5
```

##### Creating Normal distribution with Mean and Standard deviation:

```python
x = np.linspace(mean - 4 * std, mean + 4 * std, 1000)
y = norm.pdf(x, mean, std)
plt.figure(figsize=(10, 6))
plt.axvline(x=mean, color='darkgray', linestyle='--')
plt.plot(x, y, label=f"={mean},={std}")
plt.title('Normal Distribution of Breaking Strength of Gunny Bags')
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

#### 1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
x1=stats.norm.cdf(3.17,loc=mean,scale=std)
x1
```

```python
x = np.linspace(mean - 4 * std, mean + 4 * std, 1000)
y = norm.pdf(x, mean, std)
plt.figure(figsize=(10, 6))
plt.axvline(x=3.17, color='red', linestyle='-')
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability')
plt.fill_between(x, y, where=(x <= 3.17), color='red')
plt.show()
```

```python
print("Proportion of gunny bags with breaking strength less than 3.17 kg/sq.cm:", round(x1,3))
```

approximately 11.12% of the gunny bags have a breaking strength below 3.17 kg/sq.cm.

#### 2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

```python
x2=1-stats.norm.cdf(3.6,loc=mean,scale=std)
x2
```

```python
x = np.linspace(mean - 4 * std, mean + 4 * std, 1000)
y = norm.pdf(x, mean, std)
plt.figure(figsize=(10, 6))
plt.axvline(x=3.6, color='red', linestyle='-')
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability')
plt.fill_between(x, y, where=(x >=3.6), color='red')
plt.show()
```

```python
print("Proportion of of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.:", round(x2,3))
```

approximately 82.5% of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.

#### 3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
x3=stats.norm.cdf(5.5,loc=mean,scale=std)
x4=stats.norm.cdf(5,loc=mean,scale=std)
x5=x3-x4
x5
```

```python
x = np.linspace(mean - 4 * std, mean + 4 * std, 1000)
y = norm.pdf(x, mean, std)
plt.figure(figsize=(10, 6))
plt.axvline(x=5.5, color='red', linestyle='-')
plt.axvline(x=5, color='red', linestyle='-')
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability')
plt.fill_between(x, y, where=((x >= 5) & (x <= 5.5)), color='red')
plt.show()
```

```python
print("Proportion of the gunny bags having a breaking strength between 5 and 5.5 kg per sq cm. :", round(x5,3))
```

Approximately 13.1% of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.

#### 4.What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
x6=stats.norm.cdf(3,loc=mean,scale=std)
x7=stats.norm.cdf(7.5,loc=mean,scale=std)
x8=x7-x6
x9=1-x8
x9
```

```python
x = np.linspace(mean - 4 * std, mean + 4 * std, 1000)
y = norm.pdf(x, mean, std)
plt.figure(figsize=(10, 6))
plt.axvline(x=3, color='red', linestyle='-')
plt.axvline(x=7.5, color='red', linestyle='-')
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability')
plt.fill_between(x, y, where=(x <= 3), color='red')
plt.fill_between(x, y, where=(x >= 7.5), color='red')
plt.show()
```

```python
print("Proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm. :", round(x9,3))
```

Approximately 13.9% of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.
